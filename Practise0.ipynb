```python
import pandas as pd

dict = {"Fruits":["Mango", "Apple", "Banana", "Papaya"],
        "Prices":[100,150,50,35],
        "Quantity":[15,10,10,3]}

df1 = pd.DataFrame(dict)
print(df1)

df2 = df1.copy()

df2.loc[0,"Prices"]=120
df2.loc[1,"Prices"]=175
df2.loc[3,"Prices"]=35

df2.loc[0,"Quantity"]=12
df2.loc[1,"Quantity"]=15
df2.loc[3,"Quantity"]=5

print(df2)

print(df1.compare(df2) , align_axis = 0)
```

```python
import pandas as pd

data = {
    "Fruits": ["Mango", "Apple", "Banana", "Papaya"],
    "Prices": [100, 150, 50, 35],
    "Quantity": [15, 10, 10, 3]
}

df1 = pd.DataFrame(data)
print("Original DataFrame:")
print(df1)

df2 = df1.copy()

df2.loc[0, "Prices"] = 120
df2.loc[1, "Prices"] = 175
df2.loc[3, "Prices"] = 35

df2.loc[0, "Quantity"] = 12
df2.loc[1, "Quantity"] = 15
df2.loc[3, "Quantity"] = 5

print("\nModified DataFrame:")
print(df2)

df_changes = df1.compare(df2)
print("\nChanges:")
print(df_changes)
```
